# Visualization

```python
sns.pairplot(iris, hue = 'species', height = 3, aspect = 1);
plt.show();
```

```python
iris.hist(edgecolor = 'black', linewidth = 1.2, figsize = (12, 8));
plt.show();
```

```python
plt.figure(figsize=(12,8));

plt.subplot(2, 2, 1)
sns.violinplot(x = 'species', y ='sepal_length', data = iris)
plt.subplot(2, 2, 2)
sns.violinplot(x = 'species', y ='sepal_width', data = iris)
plt.subplot(2, 2, 3)
sns.violinplot(x = 'species', y ='petal_length', data = iris)
plt.subplot(2, 2, 4)
sns.violinplot(x = 'species', y ='petal_width', data = iris);

plt.show();
```

Looking at sepal_length, virginica has longest length distribution than versicolor and setosa 
Looking at sepal_width, setosa has longest width distribution than versicolor and virginica
Looking at petal_length, setosa has shortest and virginica has longest length.
Looking at petal_width, setosa has shortest and virginica has longest length.
